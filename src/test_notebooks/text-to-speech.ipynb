```python
import subprocess
import IPython.display as ipd
import os

# Run apt install espeak-ng

# ðŸ—£ï¸ Function: Generate Welsh speech using espeak-ng
def speak_welsh(text, wav_file="welsh_speech.wav"):
    # Use espeak-ng to synthesize Welsh audio
    result = subprocess.run([
        "espeak-ng",
        "-v", "cy",        # Use Welsh voice
        "-s", "140",       # Speed (words per minute, optional)
        "-w", wav_file,    # Output .wav file
        text
    ], capture_output=True, text=True)

    if result.returncode != 0:
        print("âŒ espeak-ng failed:", result.stderr)
    elif not os.path.exists(wav_file) or os.path.getsize(wav_file) == 0:
        print("âŒ Output WAV not created.")
    else:
        print(f"âœ… Speech generated: {wav_file}")
        return ipd.Audio(wav_file)

# ðŸ§ª Example Welsh sentence
welsh_text = "Sut wyt ti heddiw?"  # How are you today?

# ðŸ”Š Generate and play the audio
speak_welsh(welsh_text)
```
